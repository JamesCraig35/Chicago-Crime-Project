Crimes Across the Years

```python
dataframes = [
    df_2001, df_2002, df_2003, df_2004,
    df_2005, df_2006, df_2007, df_2008,
    df_2009, df_2010, df_2011, df_2012,
    df_2013, df_2014, df_2015, df_2016,
    df_2017, df_2018, df_2019, df_2020,
    df_2021, df_2022
]

# Loop through each DataFrame and create a bar plot
for i, df_year in enumerate(dataframes, start=2001):
    crime_types_counts = df_year['Primary Type'].value_counts()

    # Plotting the bar graph
    plt.figure(figsize=(12, 6))
    crime_types_counts.plot(kind='bar')
    plt.title(f'Distribution of Crime Types in Chicago ({i})')
    plt.xlabel('Crime Type')
    plt.ylabel('Number of Incidents')
    plt.xticks(rotation=45, ha='right') 
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
# Calculate the total number of crimes for each year
total_crimes_per_year = [df_year.shape[0] for df_year in dataframes]

# Plotting the trend
plt.figure(figsize=(12, 6))
plt.plot(range(2001, 2023), total_crimes_per_year, marker='o', linestyle='-', color='b')
plt.title('Total Number of Crimes in Chicago (2001-2022)')
plt.xlabel('Year')
plt.ylabel('Total Number of Crimes')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns

# color palette
sns.set_palette("husl")

# List of DataFrames for each year
dataframes = [
    df_2001, df_2002, df_2003, df_2004,
    df_2005, df_2006, df_2007, df_2008,
    df_2009, df_2010, df_2011, df_2012,
    df_2013, df_2014, df_2015, df_2016,
    df_2017, df_2018, df_2019, df_2020,
    df_2021, df_2022
]

# Concatenate all dataframes into one dataframe
all_data = pd.concat(dataframes)

# Convert 'Date' column to datetime format
all_data['Date'] = pd.to_datetime(all_data['Date'], errors='coerce')

# Ensure 'Date' column is in datetime format
all_data = all_data.dropna(subset=['Date'])

# Create a DataFrame to store crime trends for each type
crime_trends = all_data.groupby(['Date', 'Primary Type']).size().unstack(fill_value=0)

# Select a subset of crime types for better visualization
selected_crime_types = crime_trends.sum().nlargest(10).index

# Plot individual crime trends for the selected crime types using Seaborn
plt.figure(figsize=(15, 8))
sns.lineplot(data=crime_trends[selected_crime_types], dashes=False, markers=True)

plt.title('Individual Crime Trends in Chicago (2001-2022)')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend(title='Crime Type', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

Comparing AM vs. PM Rush Hour

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming you have a large DataFrame named all_data
# Replace 'all_data' with the actual name of your DataFrame if it's different

# Convert 'Date' column to datetime format
all_data['Date'] = pd.to_datetime(all_data['Date'], errors='coerce')

# Extract the hour from the 'Date' column
all_data['Hour'] = all_data['Date'].dt.hour

# Categorize the hours into AM and PM rush hours
all_data['Rush Hour'] = pd.cut(all_data['Hour'], bins=[0, 9, 17, 23], labels=['Late Night', 'AM Rush Hour', 'PM Rush Hour'], include_lowest=True)

# Create a countplot using Seaborn to visualize crime distribution during AM and PM rush hours
plt.figure(figsize=(10, 6))
sns.countplot(x='Rush Hour', data=all_data, palette='viridis')
plt.title('Number of Crimes during AM and PM Rush Hours')
plt.xlabel('Rush Hour')
plt.ylabel('Number of Crimes')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert 'Date' column to datetime format
all_data['Date'] = pd.to_datetime(all_data['Date'], errors='coerce')

# Extract the hour from the 'Date' column
all_data['Hour'] = all_data['Date'].dt.hour

# Filter data for AM rush hour (e.g., 6 AM to 9 AM)
am_rush_hour_data = all_data[(all_data['Hour'] >= 6) & (all_data['Hour'] < 9)]

# Get the top 5 most common crimes during AM rush hour
top_crimes_am_rush_hour = am_rush_hour_data['Primary Type'].value_counts().nlargest(5)

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x=top_crimes_am_rush_hour.index, y=top_crimes_am_rush_hour.values, palette='viridis')
plt.title('Top 5 Most Common Crimes during AM Rush Hour')
plt.xlabel('Crime Type')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
